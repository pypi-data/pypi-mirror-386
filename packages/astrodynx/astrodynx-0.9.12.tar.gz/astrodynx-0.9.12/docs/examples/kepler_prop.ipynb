# Kepler Propagation

```python
%load_ext autoreload
%autoreload 2 

from jax import numpy as jnp
import jax
import astrodynx as adx
import matplotlib.pyplot as plt

jax.config.update("jax_enable_x64", True)
jax.devices()
```

```python
r0_vec = jnp.array([1.0, 0.0, 0.0])
v0_vec = jnp.array([0.0, 1.0, 0.0])
mu = 1.0
deltat = jnp.pi
r_vec, v_vec = adx.prop.kepler(deltat, r0_vec, v0_vec, mu)
assert jnp.allclose(r_vec, jnp.array([-1.0, 0.0, 0.0]), atol=1e-6)
assert jnp.allclose(v_vec, jnp.array([0.0, -1.0, 0.0]), atol=1e-6)
```

```python
r0_vec = jnp.array([1.0, 0.0, 0.0])
v0_vec = jnp.array([0.0, 1.0, 0.0])
mu = 1.0
ts = jnp.linspace(0, 2 * jnp.pi, 12)
r_vec, v_vec = adx.prop.kepler(ts, r0_vec, v0_vec, mu)

fig, ax = plt.subplots()
ax.plot(r_vec[:, 0], r_vec[:, 1])
ax.set_aspect("equal")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Kepler Propagation")
plt.show()
```
