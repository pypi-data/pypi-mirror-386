```python
import numpy as np
from scipy.special import erf
import matplotlib.pyplot as plt 
```

```python
def F(x, h, m, s):
    return np.sqrt(np.pi) * h * s * erf((-m+x)/s) / 2
```

```python
x = np.arange(100)
h, m, s = 1, 50, 10
y = h * np.exp(-((x-m)/s)**2)
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(15,4))
ax1.plot(x, y)
ax2.plot(x, np.cumsum(y))
ax3.plot(x, F(x, h, m, s));
```

```python
area = np.sum(y)
area
```

```python
from scipy.optimize import fsolve
```

```python
def func(s):
    return F(100, h, m, s) - F(0, h, m, s) - area
root = fsolve(func, 30)
root
```
