```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def phi(qv, r):
    return 3*(np.sin(qv*r) - qv*r*np.cos(qv*r))/(qv*r)**3
```

```python
r = 30
qv = np.linspace(0.001, 0.5, 100)
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(14,7))
y = phi(qv,r)**2
ax1.plot(qv, y)
ax1t = ax1.twinx()
ax1t.plot(qv, np.sin(qv*r), ':', color='C1')
ax2.set_yscale('log')
ax2.plot(qv, y)
for ax in [ax1, ax2]:
    ymin, ymax = ax.get_ylim()
    ax.set_ylim(ymin, ymax)
    for s in [1, 1.5]:
        x = s*np.pi/r
        ax.plot([x, x], [ymin, ymax])
```
