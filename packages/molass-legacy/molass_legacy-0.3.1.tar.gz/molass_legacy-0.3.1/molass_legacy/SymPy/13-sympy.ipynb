```python
from sympy import *
x, L ,x0, k, b, y = symbols('x, L ,x0, k, b, y')
init_printing(use_unicode=True)
```

```python
solve(L/(1 + exp(-k*(x-x0))) + b - y, x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(y, L ,x0, k, b):
    return L/(1 + np.exp(-k*(x-x0))) + b
```

```python
def sigmoid_inv_sympy(y, L ,x0, k, b):
    return 1/k * np.log((-b+y)*np.exp(k*x0)/(L+b-y))
```

```python
def sigmoid_inv(y, L ,x0, k, b):
    return -np.log(L/(y - b) - 1)/k + x0
```

```python
x = np.linspace(-10, 10, 100)
y = sigmoid(x, 1, 0, 1, 0)
z = sigmoid_inv_sympy(y, 1, 0, 1, 0)
w = sigmoid_inv(y, 1, 0, 1, 0)
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8,3))
ax1.plot(x, y)
ax2.plot(y, z)
ax2.plot(y, w, ":")
ax1.plot(z, y, ":")
ax1.plot(w, y, ":");
```
