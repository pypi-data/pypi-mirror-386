```python
from sympy import symbols, simplify, solve
K, l, C = symbols("K, lambda, C", positive=True)
sol = solve([K - 1/(1 + 2*l - C*l**2)], l)
sol
```

```python
import numpy as np
import matplotlib.pyplot as plt
c = 2
x = np.linspace(0, 1, 100)
fig, (ax1, ax2) = plt.subplots(ncols=2,figsize=(10,4))
ax1.plot(1/(1 + 2*x - c*x**2), x)
ax2.plot(x, 1/c - np.sqrt(c*x - c + x)/(c*np.sqrt(x)))
ax2.plot(x, 1/c + np.sqrt(c*x - c + x)/(c*np.sqrt(x)))
ax2.plot(1/(1 + 2*x - c*x**2), x, ":");
```
