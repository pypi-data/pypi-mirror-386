# Analyze Uperf Logs

```python
# S3_LOGS_URLS = ['412URL', '413URL'] - for comparing 2 artifacts 
S3_LOGS_URLS = ['']
# Compare versions/ block size
versions = ['412', '413']
block_size = '64 1th'
```

# Prerequsite

```python

# SET log level
# logging
import logging
logger = logging.getLogger()
logger.setLevel(logging.WARN)
logging.info("Prerequisite")

# extract credentials
import configparser
config = configparser.ConfigParser()
config.read('logs.conf')
username = config.get("data", 'user') 
password = config.get("data", 'password')


# Update fedora
!dnf update -y

from benchmark_runner.jupyterlab.templates.logs_operations.logs_operations import LogsOperations
from benchmark_runner.jupyterlab.templates.analyze_workloads.analyze_uperf_logs.analyze_uperf_logs import AnalyzeUperfLogs


logs_operations_obj = []
analyze_uperf_logs_obj = []

#Logs operations
for i, S3_LOGS_URL in enumerate(S3_LOGS_URLS):
    logs_operations = LogsOperations(s3_logs_url=S3_LOGS_URL)
    logs_operations_obj.append(logs_operations)
    
# Logs cleanup
logs_operations_obj[0].cleanup()

for i, S3_LOGS_URL in enumerate(S3_LOGS_URLS):
    analyze_uperf_logs = AnalyzeUperfLogs(s3_logs_url=S3_LOGS_URL)
    analyze_uperf_logs_obj.append(analyze_uperf_logs)

# Output clean output display
analyze_uperf_logs_obj[0].clear_output()
```

## Download S3 logs

```python
for i, S3_LOGS_URL in enumerate(S3_LOGS_URLS):
    logs_operations_obj[i].download_s3_logs(username=username, password=password)
```

## Untar logs & chmod 

```python
for i, S3_LOGS_URL in enumerate(S3_LOGS_URLS):
    logs_operations_obj[i].untar_and_chmod_logs()
```

# Get Uperf log file

```python
workload_log_files=[]
for i, S3_LOGS_URL in enumerate(S3_LOGS_URLS):
    workload_log_files.append(analyze_uperf_logs_obj[i].get_uperf_log_file())
```

# Extract Uperf result

```python
uperf_results=[]
for i, S3_LOGS_URL in enumerate(S3_LOGS_URLS):
    uperf_results.append(analyze_uperf_logs_obj[i].extract_uperf_result(workload_log_file=workload_log_files[i]))
```

# Display Uperf result

```python
for i, S3_LOGS_URL in enumerate(S3_LOGS_URLS):
    analyze_uperf_logs_obj[i].display_uperf_result(uperf_result=uperf_results[i], workload_log_file=workload_log_files[i])
```

# Comapare results

```python
if len(S3_LOGS_URLS)>1:
    analyze_uperf_logs_obj[0].compare_results(result1=uperf_results[0], result2=uperf_results[1], legend_label1=versions[0], legend_label2=versions[1], block_size=block_size)
```
